```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# Create X and y 
y = df['churn']
X = df[['customer service calls', 'international plan', 'voice mail plan', 
                      'total intl calls','total day charge','total day minutes']]

# Split into training and test sets
SEED = 1
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=SEED)
```

```python
# Map 'yes' to 1 and 'no' to 0 for both columns
X['international plan'] = X['international plan'].map({'yes': 1, 'no': 0})
X['voice mail plan'] = X['voice mail plan'].map({'yes': 1, 'no': 0})


X_train['international plan'] = X_train['international plan'].map({'yes': 1, 'no': 0})
X_train['voice mail plan'] = X_train['voice mail plan'].map({'yes': 1, 'no': 0})

X_test['international plan'] = X_test['international plan'].map({'yes': 1, 'no': 0})
X_test['voice mail plan'] = X_test['voice mail plan'].map({'yes': 1, 'no': 0})
```

```python
print(X.dtypes)
```

```python
# Make predictions using test set 
y_pred = dt.predict(X_test)

# Check the AUC of predictions
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

# Identify optimal tree length 



```python
# Identify the optimal tree depth for given data
max_depths = list(range(1, 33))
train_results = []
test_results = []
for max_depth in max_depths:
    dt = DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=42)
    dt.fit(X_train, y_train)
    train_pred = dt.predict(X_train)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    # Add auc score to previous train results
    train_results.append(roc_auc)
    y_pred = dt.predict(X_test)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    # Add auc score to previous test results
    test_results.append(roc_auc)

plt.figure(figsize=(12,6))
plt.plot(max_depths, train_results, 'b', label='Train AUC')
plt.plot(max_depths, test_results, 'r', label='Test AUC')
plt.ylabel('AUC score')
plt.xlabel('Tree depth')
plt.legend()
plt.show()
```

Summary

- The model performs best with tree depths between 4 and 6.
- The test AUC improves initially (up to a depth of ~4â€“6) but plateaus and fluctuates with deeper trees, showing no additional improvement.
- A large gap between Train AUC and Test AUC for deeper trees highlights overfitting, where the model memorizes training data instead of generalizing.

# Min Sample Split

```python
# Define a range of values for min_samples_split
min_samples_splits = np.linspace(0.1, 1.0, 10, endpoint=True) * len(X_train)
min_samples_splits = min_samples_splits.astype(int)

train_results = []
test_results = []

for min_samples_split in min_samples_splits:
    # Initialize the Decision Tree Classifier
    dt = DecisionTreeClassifier(criterion='entropy', min_samples_split=min_samples_split, random_state=42)
    dt.fit(X_train, y_train)
    
    # Predict probabilities for training set
    train_pred = dt.predict_proba(X_train)[:, 1]
    false_positive_rate, true_positive_rate, _ = roc_curve(y_train, train_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    train_results.append(roc_auc)
    
    # Predict probabilities for test set
    test_pred = dt.predict_proba(X_test)[:, 1]
    false_positive_rate, true_positive_rate, _ = roc_curve(y_test, test_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    test_results.append(roc_auc)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(min_samples_splits, train_results, 'b', label='Train AUC')
plt.plot(min_samples_splits, test_results, 'r', label='Test AUC')
plt.xlabel('Min. Sample Splits')
plt.ylabel('AUC')
plt.legend()
plt.title('AUC vs Min. Sample Splits')
plt.show()
```

Summary

- Small min_samples_split leads to overfitting (high train AUC, low test AUC).
- Intermediate values (e.g., 500â€“1500) provide the best test AUC and generalization.
- Increasing the min_samples_split value reduces the model complexity by requiring more samples to split a node.

```python
# Calculate the optimal value for minimum sample leafs
min_samples_leafs = np.linspace(1, 5, 10, endpoint=True)  # Ensure 10 is an integer
train_results = []
test_results = []

for min_samples_leaf in min_samples_leafs:
    dt = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=int(min_samples_leaf), random_state=SEED)  # Ensure min_samples_leaf is an integer
    dt.fit(X_train, y_train)
    
    # Calculate AUC for training set
    train_pred = dt.predict(X_train)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    train_results.append(roc_auc)
    
    # Calculate AUC for test set
    y_pred = dt.predict(X_test)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    test_results.append(roc_auc)

# Plot the results
plt.figure(figsize=(12,6))
plt.plot(min_samples_leafs, train_results, 'b', label='Train AUC')
plt.plot(min_samples_leafs, test_results, 'r', label='Test AUC')
plt.ylabel('AUC score')
plt.xlabel('Min. Sample Leafs')
plt.legend()
plt.show()
```

Key Takeaways
- The optimal range appears to be around 4â€“5, where the test AUC starts to improve while the train AUC is still reasonably high.
- For very small values of min_samples_leaf (e.g., 1), the Train AUC is very high, showing the model is overfitting to the training data.
- The test values  improvement at higher values indicates better generalization as the model avoids overfitting.
